```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("internship_train.csv")
```

```python
y = df.pop("target")
train_target = np.array(y)
train_data = np.array(df)
#test = np.array(x.loc[1])
#res = y.loc[1]
#print(test)
#print("#"*20)
#print(res)
print(train_target[:5])
print(train_data[:5])
```

```python
std = train_data.std(axis=0)
train_data /= std
```

```python
train_data[:5]
```

```python
model = keras.Sequential()
model.add(Dense(64, activation="relu", input_shape=(train_data.shape[1],)))
model.add(Dense(32, activation="relu"))
model.add(Dense(1))

model.compile(loss="mse", metrics=[keras.metrics.RootMeanSquaredError()], optimizer=keras.optimizers.Adam(learning_rate=0.01))

#history = model.fit(x,y, epochs=40)

#plt.plot(history.history["loss"])
#plt.grid(True)
#plt.show()
```

```python
model.load_weights("3_q_regresion_model.h5")
```

```python
test_df = pd.read_csv("internship_hidden_test.csv")
test_df
```

```python
test_data = np.array(test_df)
test_data /= std
test_data
```

```python
y_pred = model.predict(test_data)
y_pred
```

```python
fin_df = pd.DataFrame(data=y_pred)
fin_df.columns = ['target']
fin_df
```

```python
fin_df.to_csv('model_prediction_1.csv', index=False)
```
